# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler , LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report , confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

features = ["StudyHours","Attendance","PastScore","SleepHours"]

scaler = StandardScaler()

df_scaled = df.copy()

df_scaled[features] = scaler.fit_transform(df[features])

x = df_scaled[features]
y = df_scaled["Passed"]

x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2 , random_state=42)

model = LogisticRegression()

model.fit(x_train,y_train)

y_pred = model.predict(x_test)

print("Classification Report")
print(classification_report(y_test , y_pred))

conf_matrix = confusion_matrix(y_test , y_pred)

plt.figure(figsize=(6, 4))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues",
            xticklabels=["Fail", "Pass"],  # FIX: Used comma for list
            yticklabels=["Fail", "Pass"])  # FIX: Used comma for list

plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.tight_layout()
plt.show()

print("\n------------- Predict Your Result -------------")
try:
    # Get user input
    study_hours = float(input("Enter study hours: "))
    attendance = float(input("Enter attendance: "))
    past_score = float(input("Enter past score: "))
    sleep_hour = float(input("Enter sleep hours: "))

    # Create a DataFrame from user input
    user_input_df = pd.DataFrame([{
        "StudyHours": study_hours,  # FIX: Corrected column name typo
        "Attendance": attendance,
        "PastScore": past_score,
        "SleepHours": sleep_hour
    }])

    # Scale the user input using the same scaler
    user_input_scaled = scaler.transform(user_input_df)

    # Make the prediction
    prediction = model.predict(user_input_scaled)[0]

    # Display the result
    result = "Pass" if prediction == 1 else "Fail"
    print(f"\n---> The predicted result is: {result}") # FIX: Added print statement

# FIX: Corrected indentation of the 'except' block
except Exception as e:
    print(f"An error occurred: {e}")
```
